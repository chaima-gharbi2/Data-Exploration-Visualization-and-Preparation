# Data Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler, StandardScaler

```

```python
# Scatter plot: Blood glucose vs. Blood pressure, differentiating diabetic and non-diabetic individuals
plt.figure(figsize=(8,6))
sns.scatterplot(data=file, x="PlGlcConc", y="BloodP", hue="HasDiabetes", palette="coolwarm")
plt.title("Blood Glucose vs. Blood Pressure")
plt.xlabel("Glucose Level")
plt.ylabel("Blood Pressure")
plt.show()

# Count the number of zero values in each column
zero_values = (file == 0).sum()
print("\nNumber of Zero Values per Column:\n", zero_values)

# Plot histograms for all attributes to visualize distributions
file.hist(figsize=(12,10), bins=20)
plt.show()
```

#  Data Preparation
a) Data Cleaning

```python
# Replace missing values in the SkinThick attribute with the median
file["SkinThick"] = file["SkinThick"].replace(0, np.nan)
median_skinthick = file["SkinThick"].median()
file["SkinThick"] = file["SkinThick"].fillna(median_skinthick)

# Verify the updated median value
print("\nNew Median for SkinThick:", file["SkinThick"].median())
```

b)Data Normalization (Min-Max Scaling)

```python
scaler = MinMaxScaler()
file_scaled = file.copy()
file_scaled.iloc[:, :-1] = scaler.fit_transform(file.iloc[:, :-1])

# Display normalized data
print("\nData After Min-Max Scaling:\n", file_scaled.head())
```

c) Data Standardization (Z-score Normalization)

```python
standard_scaler = StandardScaler()
file_standardized = file.copy()
file_standardized.iloc[:, :-1] = standard_scaler.fit_transform(file.iloc[:, :-1])

# Display standardized data
print("\nData After Standardization:\n", file_standardized.head())
```
